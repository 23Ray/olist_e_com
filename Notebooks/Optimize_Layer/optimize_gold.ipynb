```python
%sql
USE CATALOG olist_db;
USE SCHEMA gold;

OPTIMIZE daily_sales
ZORDER BY (order_date);

OPTIMIZE monthly_sales
ZORDER BY (order_month);

OPTIMIZE customer_ltv
ZORDER BY (customer_id);

OPTIMIZE seller_performance
ZORDER BY (seller_id);

OPTIMIZE seller_state_revenue
ZORDER BY (seller_state);

OPTIMIZE state_wise_revenue
ZORDER BY (customer_state);

OPTIMIZE category_performance
ZORDER BY (product_category_name_english);

OPTIMIZE top_products
ZORDER BY (product_id);

OPTIMIZE delivery_vs_reviews
ZORDER BY (review_score);

OPTIMIZE payment_method_stats
ZORDER BY (payment_type);

OPTIMIZE review_distribution
ZORDER BY (review_score);
```

```python
catalog_name = 'olist_db'
schema_name = 'gold'

gold_tables = ['category_performance', 'customer_ltv', 'daily_sales', 'delivery_performance', 'delivery_vs_reviews', 'monthly_sales', 'payment_method_stats', 'review_distribution', 'seller_performance', 'seller_state_revenue', 'state_wise_revenue', 'top_products']

for name in gold_tables:
    table_name = catalog_name + "." + schema_name + "." + name
    spark.sql(f"""
                VACUUM {table_name} RETAIN 720 HOURS
              """)
```
