```python
%sql
SELECT geolocation_zip_code_prefix, COUNT(*) FROM olist_db.silver.geolocation
GROUP BY geolocation_zip_code_prefix
HAVING COUNT(*) > 1;
```

```python
%sql
SELECT * FROM olist_db.silver.geolocation G
INNER JOIN olist_db.silver.customers c
ON G.geolocation_zip_code_prefix = c.customer_zip_code_prefix
```
